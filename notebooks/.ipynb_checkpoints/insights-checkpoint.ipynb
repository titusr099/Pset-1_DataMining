```python
import os
import snowflake.connector
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Conectar a Snowflake
conn = snowflake.connector.connect(
    user=os.getenv("SNOWFLAKE_USER"),
    password=os.getenv("SNOWFLAKE_PASSWORD"),
    account=os.getenv("SNOWFLAKE_ACCOUNT"),
    warehouse=os.getenv("SNOWFLAKE_WAREHOUSE"),
    database=os.getenv("SNOWFLAKE_DATABASE"),
    schema="GOLD"
)
```

```python
# Ejecutar una consulta de prueba
cursor = conn.cursor()
cursor.execute("SELECT CURRENT_VERSION()")
version = cursor.fetchone()
print(f"Conectado a Snowflake. Versión: {version[0]}")
```

```python
cursor.execute("SHOW DATABASES")
databases = cursor.fetchall()
for db in databases:
    print(db[1])
```

```python
cursor.execute("SHOW SCHEMAS IN INSTACART_DB")
schemas = cursor.fetchall()
for schema in schemas:
    print(schema[1])
```

```python
cursor.execute("SHOW TABLES IN INSTACART_DB.GOLD")
tables = cursor.fetchall()
for table in tables:
    print(table[1])
```

```python
cursor.execute("USE DATABASE INSTACART_DB")
cursor.execute("USE SCHEMA GOLD")
```

```python
# Función para ejecutar consultas en Snowflake y traer resultados en un DataFrame
def fetch_query(query):
    with conn.cursor() as cursor:
        cursor.execute(query)
        df = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])
    return df
```

```python
# Comportamiento de compra según día de la semana
```

```python
query = """
SELECT "order_dow", "order_dow_name", COUNT("order_id") AS "total_orders"
FROM "FactOrderProducts" f
JOIN "DimTime" t ON f."time_key" = t."time_key"
GROUP BY "order_dow", "order_dow_name"
ORDER BY "order_dow";
"""
df_dow = fetch_query(query)
```

```python
df_dow
```

```python
# Crear la figura
plt.figure(figsize=(10, 6))

# Gráfico de barras con Seaborn (ajustando `hue` para evitar la advertencia)
sns.barplot(
    data=df_dow,
    x="order_dow_name",
    y="total_orders",
    hue="order_dow_name",  # Se usa `hue` para evitar la advertencia
    palette="Blues_r",
)

# Etiquetas y título
plt.xlabel("Día de la Semana", fontsize=12)
plt.ylabel("Total de Órdenes", fontsize=12)
plt.title("Distribución de Órdenes por Día de la Semana", fontsize=14, fontweight="bold")

# Rotar etiquetas del eje X para mayor claridad
plt.xticks(rotation=45)

# Agregar línea de cuadrícula horizontal
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostrar el gráfico
plt.show()
```

Análisis:

1. Domingo es el día con más órdenes, lo que indica que muchas personas hacen compras en ese día, probablemente para abastecerse para la semana.
2. Menor actividad entre miércoles y jueves, sugiriendo que estos días tienen menos tráfico de compras.
3. Los fines de semana tienen más órdenes en comparación con los días laborales, lo que puede estar relacionado con la disponibilidad de tiempo de los clientes para hacer pedidos.

```python
# Comportamiento de compra según hora del día
```

```python
query = """
SELECT "order_hour_of_day", COUNT("order_id") AS "total_orders"
FROM "FactOrderProducts" f
JOIN "DimTime" t ON f."time_key" = t."time_key"
GROUP BY "order_hour_of_day"
ORDER BY "order_hour_of_day";
"""
df_hour = fetch_query(query)
```

```python
df_hour
```

```python
# Crear la figura
plt.figure(figsize=(12, 6))

# Graficar los datos
sns.barplot(x=df_hour["order_hour_of_day"], y=df_hour["total_orders"], color="royalblue")

# Configuración del gráfico
plt.xlabel("Hora del Día", fontsize=12)
plt.ylabel("Total de Órdenes", fontsize=12)
plt.title("Distribución de Órdenes por Hora del Día", fontsize=14, fontweight="bold")
plt.xticks(range(0, 24))
plt.grid(axis='y', linestyle="--", alpha=0.6)


plt.show()
```

Analisis:

1.	Horas Pico (9 AM - 4 PM)
    - Se observa un fuerte incremento en el número de órdenes a partir de las 7 AM.
	- El punto máximo ocurre entre las 9 AM y 4 PM, con el mayor volumen de pedidos.
	- Este horario puede representar un patrón de compra vinculado a las rutinas matutinas, como realizar pedidos antes del almuerzo.

2.	Disminución en la Tarde-Noche
	- Después de las 4 PM, la cantidad de órdenes empieza a disminuir gradualmente.
	- La caída más pronunciada se da a partir de las 6 PM, lo que sugiere que la mayoría de los pedidos ya han sido realizados en el día.

3.	Madrugada con Menor Actividad
	- Entre la 1 AM y las 6 AM hay una actividad mínima de pedidos.
	- Esto indica que la mayoría de los clientes no realizan compras en este horario.

```python
# Distribución de las órdenes hechas por los clientes
```

```python
query = """
SELECT "user_key", COUNT(DISTINCT "order_id") AS "total_orders"
FROM "FactOrderProducts"
GROUP BY "user_key"
"""
df_user_orders = fetch_query(query)
```

```python
df_user_orders
```

```python
# Crear la figura
plt.figure(figsize=(12, 6))

# Graficar la distribución de órdenes por usuario
sns.histplot(df_user_orders["total_orders"], bins=50, color="royalblue", alpha=0.7, kde=False)

# Configuración del gráfico
plt.xlabel("Órdenes por Usuario", fontsize=12)
plt.ylabel("Cantidad de Usuarios", fontsize=12)
plt.title("Distribución de Órdenes por Cliente", fontsize=14, fontweight="bold")
plt.grid(axis='y', linestyle="--", alpha=0.6)


plt.show()
```

Análisis:

1.	Distribución Asimétrica
    - La mayoría de los usuarios realizan entre 1 y 5 órdenes.
    - A medida que aumenta el número de órdenes por usuario, la frecuencia de usuarios disminuye drásticamente.
  
2. Usuarios de una Sola Orden
    - Se observa un pico alto en 1 orden, indicando que muchos usuarios solo compraron una vez.
    - Este fenómeno podría deberse a clientes ocasionales o que no encontraron valor en repetir la compra.
      
3. Clientes Leales y Recurrentes
    - Aunque en menor cantidad, algunos usuarios han realizado más de 10 órdenes, lo que representa a los clientes fieles de la tienda.

```python
# Top 20 artículos que se vuelven a pedir con más frecuencia
```

```python
query = """
SELECT p."product_name", COUNT(f."product_key") AS "reorder_count"
FROM "FactOrderProducts" f
JOIN "DimProduct" p ON f."product_key" = p."product_key"
WHERE f."reordered" = 1
GROUP BY p."product_name"
LIMIT 20;
"""
df_top_reorder = fetch_query(query)
```

```python
df_top_reorder
```

```python
# Crear la figura
plt.figure(figsize=(12, 6))

# Gráfico de barras con Seaborn
sns.barplot(
    data=df_top_reorder,
    x="reorder_count",
    y="product_name",
    color="royalblue"
)

# Configuración del gráfico
plt.xlabel("Reordenaciones", fontsize=12)
plt.ylabel("Producto", fontsize=12)
plt.title("Top 20 Productos Más Reordenados", fontsize=14, fontweight="bold")

# Invertir el eje Y para mejor visualización
plt.gca().invert_yaxis()

plt.show()
```

Análisis:

1. Ciertos productos son reordenados con mucha más frecuencia que otros.
    - Los productos en los primeros lugares como el Organic Avocado, Honeycrisp Apple y Organic Whole String Cheese destacan significativamente.

2. Tendencia hacia productos orgánicos y saludables.
    - Muchos de los productos más reordenados son frutas, lácteos y alimentos orgánicos.
    - Esto sugiere que los clientes de esta tienda tienen una preferencia marcada por productos frescos y saludables.

3. Algunos productos tienen una frecuencia muy baja de reordenación.
    - Productos como Enchilada Black Bean Vegetable y Parchment Paper tienen menos reordenaciones.
    - Esto podría indicar que son compras ocasionales en lugar de productos esenciales.

```python
# Proporción de pedidos que se vuelven a pedir para cada producto
```

```python
query = """
SELECT p."product_name", 
       COUNT(f."product_key") AS "total_orders",
       SUM(CASE WHEN f."reordered" = 1 THEN 1 ELSE 0 END) AS "reordered_orders",
       SUM(CASE WHEN f."reordered" = 1 THEN 1 ELSE 0 END) / COUNT(f."product_key") AS "reorder_rate"
FROM "FactOrderProducts" f
JOIN "DimProduct" p ON f."product_key" = p."product_key"
GROUP BY p."product_name"
LIMIT 20;
"""
df_reorder_rate = fetch_query(query)
```

```python
df_reorder_rate
```

```python
# Crear la figura
plt.figure(figsize=(12, 6))

# Gráfico de barras con Seaborn
sns.barplot(
    data=df_reorder_rate,
    x="reorder_rate",
    y="product_name",
    color="royalblue"
)

# Configuración del gráfico
plt.xlabel("Proporción de Reordenaciones", fontsize=12)
plt.ylabel("Producto", fontsize=12)
plt.title("Top 20 Productos con Mayor Reordenación", fontsize=14, fontweight="bold")

# Mejorar la visualización del eje Y
plt.gca().invert_yaxis()


plt.show()
```

Análisis:

1. Diferencia entre frecuencia de compra y tasa de reordenación

    - Aunque algunos productos se compran con mucha frecuencia, no necesariamente tienen la tasa de reordenación más alta.
    - En este caso, los productos con más alta proporción de reordenación no siempre son los más comprados.

3. Patrón de productos con alta reordenación
    - Se observa una tendencia hacia productos fáciles de consumir regularmente, como:
        - Snacks (Goldfish Crackers, Traditional Snack Mix)
        - Lácteos (Half And Half Cream, Vanilla Almond Breeze Almond Milk, Danimals Smoothie)
        - Productos Orgánicos (Organic Cilantro, Organic Whole White Mushrooms)
     
4. Algunos productos tienen más del 50% de reordenación
    - Lo que indica que la mayoría de las personas que los compran, vuelven a hacerlo.

```python
# Proporción de productos pedidos que se vuelven a pedir para cada cliente.
```

```python
query = """
SELECT "user_key",
       COUNT("product_key") AS "total_products",
       SUM(CASE WHEN "reordered" = 1 THEN 1 ELSE 0 END) AS "reordered_products",
       SUM(CASE WHEN "reordered" = 1 THEN 1 ELSE 0 END) / COUNT("product_key") AS "reorder_ratio"
FROM "FactOrderProducts"
GROUP BY "user_key"
"""
df_user_reorder = fetch_query(query)
```

```python
df_user_reorder
```

```python
# Configuración del tamaño de la figura
plt.figure(figsize=(10, 6))

# Histograma con Seaborn
sns.histplot(df_user_reorder["reorder_ratio"], bins=50, kde=True, color='royalblue', alpha=0.7)

# Configuración de etiquetas y título
plt.xlabel("Proporción de Productos Reordenados", fontsize=12)
plt.ylabel("Cantidad de Clientes", fontsize=12)
plt.title("Distribución de la Proporción de Productos Reordenados por Cliente", fontsize=14, fontweight="bold")

# Líneas de cuadrícula
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

Análisis:

1.	Distribución Bimodal
    - Se observa un pico alto en 0.0, lo que indica que hay muchos clientes que nunca reordenan productos.
    - También hay un pico en 1.0, lo que significa que ciertos clientes reordenan el 100% de sus compras, lo que sugiere patrones de compra altamente repetitivos.
      
2.	Patrones Intermedios
    - Entre 0.2 y 0.8 hay una distribución uniforme con algunas concentraciones en ciertos valores.
    - Esto indica que la mayoría de los clientes reordenan algunos productos pero también compran nuevos.
      
3.	Segmentación de Clientes
    - Compradores exploradores (0.0 - 0.2): Compran productos nuevos en la mayoría de sus órdenes.
    - Compradores balanceados (0.3 - 0.7): Tienen una mezcla de productos nuevos y reordenaciones.
    - Compradores leales (0.8 - 1.0): Recompran los mismos productos de manera consistente.

```python
# Top 20 artículos que la gente pone primero en el carrito
```

```python
query = """
SELECT p."product_name", COUNT(f."product_key") AS "first_added_count"
FROM "FactOrderProducts" f
JOIN "DimProduct" p ON f."product_key" = p."product_key"
WHERE f."add_to_cart_order" = 1
GROUP BY p."product_name"
LIMIT 20;
"""
df_first_cart = fetch_query(query)
```

```python
# Configurar la figura
plt.figure(figsize=(12, 6))

# Crear gráfico de barras horizontal con Seaborn
sns.barplot(
    data=df_first_cart,
    y="product_name",
    x="first_added_count",
    hue="product_name",
    legend=False,
    palette="Blues_r"
)

# Configuración del gráfico
plt.xlabel("Veces Primero en el Carrito", fontsize=12)
plt.ylabel("Producto", fontsize=12)
plt.title("Top 20 Productos Agregados Primero al Carrito", fontsize=14, fontweight="bold")

# Mostrar el gráfico
plt.show()
```

Análisis:

1.	Large Lemon es el producto más agregado primero al carrito con una diferencia significativa respecto a otros productos.
2.	Organic Baby Carrots y Organic Baby Spinach también tienen una alta frecuencia, lo que indica que los consumidores tienden a agregar estos productos como sus primeras elecciones.
3.	La distribución sugiere que los productos más agregados en primer lugar suelen ser frescos y básicos, como frutas, vegetales y productos esenciales en la alimentación diaria.
4.	Algunos productos tienen una baja frecuencia, lo que indica que no suelen ser la primera elección de los compradores.
